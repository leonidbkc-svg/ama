```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Категориальные переменные
drink_type = np.random.choice(['Эспрессо', 'Латте', 'Капучино'], size=50)
size = np.random.choice(['Маленький', 'Средний', 'Большой'], size=50)
barista = np.random.choice(['Анна', 'Иван', 'Мария'], size=50)

# Числовые переменные
price = np.random.uniform(120, 350, size=50).round(0)
time_to_prepare = np.random.uniform(1, 6, size=50).round(1)
customer_rating = np.random.uniform(3.0, 5.0, size=50).round(1)

# Дата/время
date = pd.date_range(start='2025-01-01', periods=50, freq='D')

# Формируем DataFrame
data = pd.DataFrame({
    'drink_type': drink_type,
    'size': size,
    'barista': barista,
    'price': price,
    'time_to_prepare': time_to_prepare,
    'customer_rating': customer_rating,
    'date': date
})

display(data.head())
```

Цена напитка зависит от его типа и размера

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.boxplot(x='drink_type', y='price', hue='size', data=data, palette='Set3')
plt.title('Зависимость цены напитка от типа и размера')
plt.xlabel('Тип напитка')
plt.ylabel('Цена (₽)')
plt.tight_layout()
plt.show()
```
