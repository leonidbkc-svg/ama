```python
# Задание 1
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing

data = {
    'сотрудник': ['Алина', 'Борис', 'Владимир', 'Дарья', 'Егор', 'Федор', 'Галина', 'Наталья', 'Игорь', 'Юлия'],
    'отдел': ['Аналитика', 'Аналитика', 'Маркетинг', 'Аналитика', 'Маркетинг', 'Аналитика', 'Маркетинг', 'Аналитика', 'Маркетинг', 'Аналитика'],
    'проект_баллы': [85, 92, 78, np.nan, 88, 75, 95, np.nan, 82, 90],
    'выполнено_задач': [5, 7, 6, 7, 5, 6, 8, 4, 5, 7],
    'пройдено_тест': ['да', 'да', 'нет', 'да', 'да', 'нет', 'да', 'нет', 'да', 'да']
}

labels = ['с1','с2','с3','с4','с5','с6','с7','с8','с9','с10']
df_sotrudniki = pd.DataFrame(data, index=labels)
print(df_sotrudniki)

# Задание 3
print(df_sotrudniki.head(3))

# Задание 4
print(df_sotrudniki.loc[['с4', 'с5', 'с9'], ['сотрудник', 'проект_баллы']])

# Задание 5
print(df_sotrudniki[df_sotrudniki['выполнено_задач'] > 6])

# Задание 6
df_sotrudniki.loc['с6', 'проект_баллы'] = 80
print(df_sotrudniki)

# Задание 7
total_tasks = df_sotrudniki['выполнено_задач'].sum()
print(total_tasks)

# Задание 8
mean_scores = df_sotrudniki.groupby('отдел')['проект_баллы'].mean()
print(mean_scores)

# Задание 9
sorted_df = df_sotrudniki.sort_values(by=['проект_баллы', 'выполнено_задач'], ascending=[False, True])
print(sorted_df)

# Задание 10
df_sotrudniki['пройдено_тест'] = df_sotrudniki['пройдено_тест'].replace({'да': True, 'нет': False})
print(df_sotrudniki)

# Задание 11
california = fetch_california_housing()
df_california = pd.DataFrame(california.data, columns=california.feature_names)
print(df_california.head())

# Задание 12
df_california = pd.concat([df_california, pd.DataFrame(california.target, columns=['MedHouseValue'])], axis=1)
print(df_california.head())

# Задание 13
df_california.info()

# Задание 14
print(df_california.isna().sum())

# Задание 15
filtered = df_california.loc[(df_california['HouseAge'] > 50) & (df_california['Population'] > 2500)]
print(filtered)

# Задание 16
print(df_california['MedHouseValue'].max())
print(df_california['MedHouseValue'].min())
```
